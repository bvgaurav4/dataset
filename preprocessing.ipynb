```python
import pandas as pd

# Create an example DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# List of new rows to be added
new_rows = [{'A': 4, 'B': 7}, {'A': 5, 'B': 8}]

# Convert list of new rows to DataFrame
new_rows_df = pd.DataFrame(new_rows)

# Concatenate all at once
df = pd.concat([df, new_rows_df], ignore_index=True)

print(df)
```

```python
import pandas as pd
import numpy as np

# Assuming df is your original DataFrame

# Create a condition to check if 'abstract' is of type float
condition = df['abstract'].apply(lambda x: isinstance(x, float))

# Split the DataFrame based on the condition
df1 = df[condition].copy()  # Rows where 'abstract' is float
df = df[~condition].copy()  # Rows where 'abstract' is not float

# Reset the index if needed
df1.reset_index(drop=True, inplace=True)
df.reset_index(drop=True, inplace=True)

print(df1)
print(df)
```
